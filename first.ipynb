```python
# compiling

def extract(first, last, month, month_n):
    for day in range(first, last+1):
        if day < 10:
            day = str(day)
            day = '0' + day
            
        response = req.get(f'https://covid19.karnataka.gov.in/storage/pdf-files/EMB-{month}21/{day}-{month_n}-2021%20HMB%20English.pdf', stream = True)
        with open(f'{day}-{month_n}-2021%20HMB%20English.pdf', 'wb') as pdf:
#             https://covid19.karnataka.gov.in/storage/pdf-files/EMB--JUN21//28-06-2021%20HMB%20English.pdf
#             https://covid19.karnataka.gov.in/storage/pdf-files/EMB-JUL21//01-07-2021%20HMB%20English.pdf
#             https://covid19.karnataka.gov.in/storage/pdf-files/EMB-AUG21//01-08-2021%20HMB%20English.
            pdf.write(response.content)
        
        pdf_path = f'{day}-{month_n}-2021%20HMB%20English.pdf'
        dist_table = tabula.read_pdf(pdf_path, pages = 5, lattice = True)
        
        dist_table = dist_table[0]
        dist_table.columns = dist_table.iloc[1].replace('\r',' ', regex=True)
        
        dist_table = dist_table.iloc[2:32,2:12] # June: [2:32,2:11], # July: [2:32,2:12] # aug: [2:32,2:11]
        
        for col in dist_table.columns:
            if 'l.' in col: # seen in some july tables
                dist_table.drop(col, axis = 1, inplace = True)
                
#         if 'l. o' in list(dist_table.columns): # seen in some tables in July
#             dist_table.drop('l. o', axis = 1, inplace = True)
            
#         if 'Sl. o' in list(dist_table.columns): # seen in some tables in July
#             dist_table.drop('l. o', axis = 1, inplace = True)
            
        dist_table.dropna(axis = 1, inplace = True)
            
        dist_table.reset_index(inplace = True)
        dist_table.drop('index', axis = 1, inplace = True)
    
        globals()['%s%s' % (month, day)] = dist_table # naming tables for each day eg) table01 for 1st of the month
        
        path = os.path.join(r'C:\Users\Sarah Mathew\Desktop\COVID Sakhyaa\new', f'{month}')
        globals()['%s%s' % (month, day)].to_csv(os.path.join(path, f'{month}{day}.csv'), index = False)
        
        print(month, day)
#         saved = pd.read_csv(os.path.join(path, f'{month}{day}.csv'))
#         print('Saved shape: ', saved.shape)
# #         print(saved.info)
```

```python
def district_extract(district_df, district):
    
    for month in ['JUN', 'JUL', 'AUG']:
        save_path = fr'C:\Users\Sarah Mathew\Desktop\COVID Sakhyaa\new\{month}'
    
        if month == 'JUN':
            a = 28
            b = 30
            n = 6
        elif month == 'JUL':
            a = 1
            b = 31
            n = 7
        elif month == 'AUG':
            a = 1
            b = 20
            n = 8

        for date in range(a, b+1):
            if date < 10:
                date = '0'+str(date)

            df = pd.read_csv(os.path.join(save_path, f'{month}{date}.csv'))
            df.columns = district_df.columns
            
            names = difflib.get_close_matches(district, df['District Name'], n = 1) # doc: https://docs.python.org/3.4/library/difflib.html#difflib.get_close_ma
            # the above helps accomodate spelling errors/differences noticed in different documents
            #             print('close matches: ', names)
            
            for name in names: 
#                 print(name)
                dis = df[df['District Name'] == name]
                district_df = pd.concat([district_df, dis])

#             dis = df[df['District Name'] == district]
#             district_df = pd.concat([district_df, dis])
#             print(dis.shape)
#             print(date)

            
    alldates = pd.date_range(start = '28/06/2021', end = '20/08/2021')
    district_df.insert(0, 'date', alldates)

    district_df.reset_index(inplace = True)
    district_df.drop('index', axis = 1, inplace = True)  
    district_df.to_csv(f'{district}.csv', index = False)
    return district_df.shape
```

## Adding the CFR column: Case Fatality Rate

```python
new_dfs = [Bagalakote_merged, Ballari_merged, Belagavi_merged, Bengaluru_Rural_merged, Bengaluru_Urban_merged, Bidar_merged, Chamarajanagara_merged, 
           Chikkaballapura_merged, Chikkamagaluru_merged, Chitradurga_merged, Dakshina_Kannada_merged, Davanagere_merged, 
           Dharwada_merged, Gadag_merged, Hassana_merged, Haveri_merged, Kalaburagi_merged, Kodagu_merged, Kolara_merged, 
           Koppala_merged, Mandya_merged, Mysuru_merged, Raichuru_merged, Ramanagara_merged, Shivamogga_merged, Tumakuru_merged, 
           Udupi_merged, Uttara_Kannada_merged, Vijayapura_merged,Yadagiri_merged]

print(len(new_dfs))
```
